# FeathersV1 Dataset

Example of load images and classes data from NumPy files.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder

%matplotlib inline
```

### Load NumPy images and species data

```python
X_train, X_test = (
    np.load("data/numpy/train_images.npy"),
    np.load("data/numpy/test_images.npy"),
)

y_train, y_test = (
    np.load("data/numpy/train_species.npy"),
    np.load("data/numpy/test_species.npy"),
)
```

### Plot train data

```python
fig = plt.figure(figsize=(16, 5))
rows = 2
cols = 8
for i in range(0, cols * rows):
    fig.add_subplot(rows, cols, i + 1)
    plt.ylabel(y_train[i])
    plt.imshow(X_train[i])
plt.show()
```

### Encode labels

```python
enc_species = LabelEncoder()
enc_species.fit(np.load("data/numpy/set_species.npy"))

y_train, y_test = (
    enc_species.transform(y_train),
    enc_species.transform(y_test),
)
```

### Plot encoded train data

```python
fig = plt.figure(figsize=(16, 5))
rows = 2
cols = 8
for i in range(0, cols * rows):
    fig.add_subplot(rows, cols, i + 1)
    plt.ylabel(y_train[i])
    plt.imshow(X_train[i])
plt.show()
```
